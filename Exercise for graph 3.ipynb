```python
from typing import Dict, TypedDict, List
from langgraph.graph import StateGraph
```

```python
class AgentState(TypedDict):
    name: str
    age: int
    skills: List[str]
    result: str
```

```python
def node_1(state: AgentState) -> AgentState:
    """Agent that adds the name"""
    state['result'] = f"Hello, {state['name']}!"
    return state
```

```python
def node_2(state: AgentState) -> AgentState:
    """Agent that adds the age"""
    state['result'] += f" You are {str(state['age'])} years old."
    return state
```

```python
def node_3(state: AgentState) -> AgentState:
    """Agent that adds the skills"""
    skills_str = ", ".join(state['skills'])
    state['result'] += f" Your skills are: {skills_str}."
    return state
```

```python
graph = StateGraph(AgentState)
graph.add_node("name", node_1)
graph.add_node("age", node_2)
graph.add_node("skills", node_3)
graph.add_edge("name", "age")
graph.add_edge("age", "skills")
graph.set_entry_point("name")
graph.set_finish_point("skills")
```

```python
test = graph.compile()
```

```python
from IPython.display import Image, display
display(Image(test.get_graph().draw_mermaid_png()))
```

```python
final_statement = test.invoke({"name": "Teresa", "age": 28, "skills": ["Art", "Bullying"]})
```
